```python
data = pd.read_csv('AAPL.csv',parse_dates=['Date'],index_col='Date')
data
```

**retrieve only Jan 2017 stock prices**

```python
data['2017-01']
```

**find average price of apple's stock in Jan 2017**

```python
data['2017-01'].Close.mean()
```

**resampling**

```python
data.Close.resample('y').mean()
```

```python
data.Close.resample('y').sum()
```

```python
data.Close.resample('m').sum()
```

```python
%matplotlib inline
data.Close.resample('m').sum().plot()
```
